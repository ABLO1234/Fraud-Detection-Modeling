```python
## Analyse Bivariée
# Cette fonction effectue une analyse bivariée sur un DataFrame donné.
def BivariateAnalysis(df):

    # Définition des types variables
    variables_quant = df.select_dtypes(include=[float, int]).columns.tolist()
    variables_qual = df.select_dtypes(include=[object, "category"]).columns.tolist()
    
    if "isFraud" in variables_qual:
        variables_qual.remove("isFraud")

    variables_globales = variables_quant + variables_qual

    # Analyse Quantitative
    def BivariateQuant(df, variables_quant):
        for i in variables_quant:
            for j in variables_quant:
                if i != j:  # éviter doublons et auto-corrélations
                    sns.scatterplot(data=df, x=i, y=j)
                    plt.title(f"Nuage de points entre {i} et {j}")
                    plt.show()

                    coef, pvalue = stat.spearmanr(df[i], df[j])
                    print(f"Corrélation Spearman entre {i} et {j} : coef={coef:.2f}, pvalue={pvalue:.4f}")
        
        plt.figure(figsize=(10, 8))
        sns.heatmap(df[variables_quant].corr(method="spearman"), annot=True, fmt=".2f", cmap="coolwarm")
        plt.title("Matrice de Corrélation Spearman")
        plt.show()

    # Analyse Qualitative
    def BivariateQual(df, variables_qual):
        for i in variables_qual:
            for j in variables_qual:
                if i != j: 
                    table = pd.crosstab(df[i], df[j])
                    chi2, pvalue, _, _ = stat.chi2_contingency(table)
                    print(f"Chi2 entre {i} et {j} : pvalue={pvalue:.4f}")

    # Analyse Quantitatif vs Qualitatif
    def BivariateQuantQual(df, variables_qual, variables_quant):
        for qual in variables_qual:
            for quant in variables_quant:
                groups = [df[quant][df[qual] == cat] for cat in df[qual].unique()]
                if len(groups) > 1:
                    stat_kruskal, pvalue = stat.kruskal(*groups)
                    print(f"Kruskal-Wallis entre {quant} et {qual} : pvalue={pvalue:.4f}")
                

                    # Boxplot
                    plt.figure(figsize=(10, 5))
                    sns.boxplot(x=qual, y=quant, data=df)
                    plt.title(f"Boxplot de {quant} par {qual}")
                    plt.show()

    print("### Analyse Quantitative ###")
    BivariateQuant(df, variables_quant)

    print("\n### Analyse Qualitative ###")
    BivariateQual(df, variables_qual)

    print("\n### Analyse Quantitative vs Qualitative ###")
    BivariateQuantQual(df, variables_qual, variables_quant)
```
